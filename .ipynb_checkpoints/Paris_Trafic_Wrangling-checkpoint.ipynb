```python
import os
import pandas as pd
import json
import  matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from shapely.geometry import shape
```

```python
os.getcwd()
```

```python
#Open DataFrame
df=pd.read_csv("comptages.csv",sep=";")
df.head()
```

```python
#Check nan values
df.isna().sum()
```

```python
#not too many nans. We can drop them
df=df.dropna()
```

```python
#Split coordinates into long and lat
df[["latitude","longitude"]]=df.geo_point_2d.str.split(",",expand=True)

#Create ID for each location
locations=df[["longitude","latitude"]].drop_duplicates().sort_values(by="longitude").reset_index().drop(["index"],axis=1).reset_index()
df=pd.merge(df,locations,how="left" ,on=["longitude","latitude"])

#Add categorical
df.Libelle = pd.Categorical(df.Libelle)
df['road_ID'] = df.Libelle.cat.codes
```

```python
#load geoson to shapely 
df["geo_shape"]=df["geo_shape"].apply(lambda x : eval(x))
df["geo_shape"]=df["geo_shape"].apply(lambda x : shape(x))
```

```python
#Create column for road start en road end

def extract_longitude_start(x):
    return(x.coords[0][0])

def extract_latitude_start(x):
    return(x.coords[0][1])

def extract_longitude_end(x):
    return(x.coords[1][0])
                    
def extract_latitude_start(x):
    return(x.coords[1][1])


df["road_start_lon"]=df["geo_shape"].apply(extract_longitude_start)
df["road_start_lat"]=df["geo_shape"].apply(extract_latitude_start)
df["road_end_lon"]=df["geo_shape"].apply(extract_longitude_end)
df["road_end_lat"]=df["geo_shape"].apply(extract_latitude_end)
```

```python
#Correct columns types
df=df.astype({'longitude': 'float64','latitude': 'float64'})
df["Date et heure de comptage"]=pd.to_datetime(df["Date et heure de comptage"].str.split("+").str[0])
```

```python
#Rename Columns
to_rename={"Date et heure de comptage":"timestamp", "Identifiant arc": "id_arc","index" : "location_ID", "road_ID": "road_ID","Libelle":"road_name","longitude":"longitude","latitude":"latitude","geo_shape":"geo_shape","Etat arc": "road_status", "Etat trafic":"trafic_status","Débit horaire": "flow_rate","Taux d'occupation": "occupation_rate","road_start_lon":"road_start_lon","road_start_lat":"road_start_lat","road_end_lon":"road_end_lon","road_end_lat":"road_end_lat"}
df=df.rename(columns=to_rename)
#Drop unnecessary columns
df=df[list(to_rename.values())]
```

```python
#Add columns for year, month, dayof year, weekday, and hour
df["date"]=df["timestamp"].apply(lambda x : x.date())
df["year"]=df["timestamp"].apply(lambda x : x.year)
df["month"]=df["timestamp"].apply(lambda x : x.month)
df["day"]=df["timestamp"].apply(lambda x : x.day)
df["dayofyear"]=df["timestamp"].apply(lambda x : x.dayofyear)
df["weekday"]=df["timestamp"].apply(lambda x : x.weekday())
df["hour"]=df["timestamp"].apply(lambda x : x.hour)
```

```python
#Reorder columns
columns=["timestamp","date", "year","month","day","dayofyear","weekday","hour","location_ID","id_arc", "road_ID","road_name","longitude","latitude","road_start_lon","road_start_lat","road_end_lon","road_end_lat","road_status", "trafic_status","flow_rate","occupation_rate"]
df=df[columns]
```

```python
#Sort values by date
df=df.sort_values(by="timestamp")
```
